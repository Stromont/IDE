```python
# Импорт необходимых библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from IPython.display import IFrame
import datetime as dt
```

```python
solar_data = pd.read_csv('data/solarpower.csv')
solar_data
```

```python
solar_data['date']  = pd.to_datetime(solar_data['date'])
```

```python
solar_data.info()
```

```python
solar_data.shape
for i in range(0,solar_data.shape[0]):
    if solar_data['cum_power'][i] != (-1):
        continue
    solar_data['cum_power'][i] = (solar_data['cum_power'][i-1] + solar_data['cum_power'][i+1])/2
```

```python
round(solar_data['cum_power'].mean(),0)
```

Эталонный ответ  - 23848.0

```python
solar_data['day_power'] = solar_data['cum_power']
for i in range(1,solar_data.shape[0]):
    solar_data['day_power'][i] = solar_data['cum_power'][i] - solar_data['cum_power'][i-1]
solar_data['day_power'][0] = round(solar_data['day_power'][1::].mean(),0)
print(solar_data)
```

```python
solar_data['period'] = solar_data['date'].dt.to_period('M')
print(solar_data.groupby('period')['day_power'].mean())
```

```python
grouped_df = solar_data.groupby('period')['day_power'].agg('mean')
grouped_df = grouped_df.reset_index()
grouped_df['period'] = grouped_df['period'].astype(str)
print(grouped_df)
fig = px.bar(
    data_frame=grouped_df,
    x='period',  # Указываем, какой столбец использовать на оси X
    y='day_power'  # Указываем, какой столбец использовать на оси Y
)
fig.show()
```
